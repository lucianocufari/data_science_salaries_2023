#### 1.1.3.7. <a id='toc1_1_3_7_'></a>[Gráfica para visualizar nuestras variables numéricas:](#toc0_)

```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(24, 6))
cols = ['work_year', 'salary', 'salary_in_usd', 'remote_ratio']

for i, col in enumerate(cols):
    sns.histplot(x=col, data=df, kde=True, ax=axes[i])
    axes[i].set_title(f"Distribución {col}")
    
plt.tight_layout()
plt.show()
```

#### 1.1.4.1. <a id='toc1_1_4_1_'></a>[Información interesante:](#toc0_)

#### 1.1.4.2. <a id='toc1_1_4_2_'></a>[¿Cuál es el salario promedio por nivel de experiencia?](#toc0_)

```python
fig = plt.figure(figsize=(12, 8))

sns.boxplot(x='experience_level', y='salary_in_usd', data=df)
```

```python
salario_medio = df.groupby('experience_level')['salary_in_usd'].mean()
fig = plt.figure(figsize=(12, 8))
salario_medio.plot(kind='bar')
plt.title('Salario medio por nivel de experiencia')
```

#### 1.1.4.3. <a id='toc1_1_4_3_'></a>[¿Cómo ha cambiado el salario medio con el paso del tiempo (work_year)?](#toc0_)

```python
plt.figure(figsize=(12, 4))
salary_time = sns.lineplot(x='work_year', y='salary_in_usd', data=df, color='red')
salary_time.set_title('Evolución del salario en función de los años')
salary_time.set_xlabel('Años')
salary_time.set_ylabel('Salario medio') 
plt.show()
```

#### 1.1.4.4. <a id='toc1_1_4_4_'></a>[¿Cuál es la proporción de trabajo remoto entre diferentes roles de trabajo?](#toc0_)

```python
average_remote_ratio_per_job = df.groupby(["job_title"])["remote_ratio"].mean().reset_index()
average_remote_ratio_per_job = average_remote_ratio_per_job.sort_values(by="remote_ratio", ascending=True)
plt.figure(figsize=(18, 35))
sns.barplot(data = average_remote_ratio_per_job , x = 'remote_ratio', y= 'job_title', hue='remote_ratio', palette="viridis")
plt.title("Evolución del porcentaje de trabajo remoto en función del puesto de trabajo")
plt.ylabel("Ratio de trabajo remoto (en %)")
plt.xlabel("Puesto de trabajo")
plt.show()
```

#### 1.1.4.5. <a id='toc1_1_4_5_'></a>[¿Existe una relación entre el tamaño de la empresa y el salario?](#toc0_)

```python
df["company_size_num"] = df["company_size"].map({"S":1, "M":2, "L":3 } )
corr = df[["salary_in_usd", "company_size_num"]].corr()
corr
```

```python
sns.heatmap(corr, annot=True, linewidths=.5, square=True, cmap='coolwarm')
```

#### 1.1.4.6. <a id='toc1_1_4_6_'></a>[¿Cuál es el país con más trabajadores remotos?](#toc0_)

```python
remotos = df[df['remote_ratio'] == 100] # Esto filtra subconjuntos del dataframe con un condicional.
remotos_by_country = remotos.groupby('company_location').size()
plt.figure(figsize=(12, 6))
sns.barplot(x=remotos_by_country.index, y=remotos_by_country.values)
plt.title('Número de empleados que trabajan completamente de forma remota por país')
plt.xlabel('País')
plt.ylabel('Número de empleados remotos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 1.1.4.7. <a id='toc1_1_4_7_'></a>[¿Cómo ha cambiado la proporción de trabajo remoto con el paso del tiempo?](#toc0_)

```python
plt.figure(figsize=(6, 3))
sns.lineplot(x= "work_year", y="remote_ratio", data=df , color= "green")
plt.ylabel("Proporción de trabajadores remoto")
plt.xlabel("Paso del tiempo")
plt.show()
```

#### 1.1.4.8. <a id='toc1_1_4_8_'></a>[¿Cuál es el país con los salarios promedio más altos en USD?](#toc0_)

```python
plt.figure(figsize=(16,8))
order = df.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False).index.values
sns.barplot(data = df,
            x='employee_residence',
            y='salary_in_usd',
            order=order)
plt.title('Países más salarios más altos en USD por país')
plt.xticks(rotation=90)
plt.ylabel('Salario')
plt.xlabel('País')
plt.show()
```

#### 1.1.4.9. <a id='toc1_1_4_9_'></a>[¿Cómo se distribuyen los roles de trabajo en las diferentes categorías de experiencia?](#toc0_)

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 12))

# SE
ax1 = axes[0, 0]
sns.countplot(x='job_title', data=df[df['experience_level'] == 'SE'], ax=ax1)
ax1.set_title('Distribution of Job Roles for SE')
ax1.set_xlabel('Job Title')
ax1.set_ylabel('Count')
ax1.tick_params(axis='x', rotation=90)

# MI
ax2 = axes[0, 1]
sns.countplot(x='job_title', data=df[df['experience_level'] == 'MI'], ax=ax2)
ax2.set_title('Distribution of Job Roles for MI')
ax2.set_xlabel('Job Title')
ax2.set_ylabel('Count')
ax2.tick_params(axis='x', rotation=90)

# EN
ax3 = axes[1, 0]
sns.countplot(x='job_title', data=df[df['experience_level'] == 'EN'], ax=ax3)
ax3.set_title('Distribution of Job Roles for EN')
ax3.set_xlabel('Job Title')
ax3.set_ylabel('Count')
ax3.tick_params(axis='x', rotation=90)

# EX
ax4 = axes[1, 1]
sns.countplot(x='job_title', data=df[df['experience_level'] == 'EX'], ax=ax4)
ax4.set_title('Distribution of Job Roles for EX')
ax4.set_xlabel('Job Title')
ax4.set_ylabel('Count')
ax4.tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

#### 1.1.4.10. <a id='toc1_1_4_10_'></a>[¿Cómo varía el salario promedio entre los diferentes tipos de empleo, fulltime, partime ect?](#toc0_)

```python
avg_salary_by_employment = df.groupby('employment_type')['salary_in_usd'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_salary_by_employment.sort_values('salary_in_usd', ascending=False), 
            x='employment_type', y='salary_in_usd', hue= 'employment_type', palette='cool', edgecolor='black')
plt.title('Salario medio por tipo de empleo')
plt.xlabel('Tipo de empleo')
plt.ylabel('Salario medio (USD)')
plt.show()
```

#### 1.1.4.11. <a id='toc1_1_4_11_'></a>[Top 10 de los empleos mas demandados:](#toc0_)

```python
top10_empleos = df.job_title.value_counts()[:10].index.tolist()
plt.figure(figsize=(8, 8))
sns.countplot(data=df[df.job_title.isin(top10_empleos)], y='job_title',hue='job_title', palette='Set3')
plt.title('Top 10 de empleos en Data')
plt.xlabel('Cantidad de empleados')
plt.ylabel('Título del empleo')
plt.show()
```

#### 1.1.4.12. <a id='toc1_1_4_12_'></a>[ Top 10 de los empleos mas cotizados:](#toc0_)

```python
top10_cotizados = df.groupby('job_title')['salary_in_usd'].mean().reset_index()
top10_cotizados = top10_cotizados.sort_values(by='salary_in_usd', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(data=top10_cotizados, y='job_title', x='salary_in_usd', hue='job_title', palette='Set3')
plt.title('Top 10 empleos más cotizados')
plt.xlabel('Salario medio (USD)')
plt.ylabel('Título del empleo')
plt.show()
```

#### 1.1.4.13. <a id='toc1_1_4_13_'></a>[ Top 10 de empleos con posibilidad de trabajo remoto:](#toc0_)

```python
top10_remote_jobs = df[df['remote_ratio'] ==100].groupby('job_title')['remote_ratio'].count().nlargest(10).index.tolist()
plt.figure(figsize=(8, 8))
sns.countplot(data=df[df['job_title'].isin(top10_remote_jobs)], y='job_title', hue='job_title', palette='Set3')
plt.title('Top 10 trabajos remotos')
plt.xlabel('Número de empleados')
plt.ylabel('Título del empleo')
plt.show()
```
